```python
#Tranforming dataframe by running two loops & columns as items
trans = []
for i in range (0, 7500):
    trans.append([str(df.values[i,j]) for j in range (0,20)])
    
#Transform list into arrays   
TE = TransactionEncoder()
array = TE.fit(trans).transform(trans)
```

```python
df.to_csv('D212Task3.csv')
```

```python
#Logical rows
#Columns are purchased merchandise
cleaned_df = pd.DataFrame(array, columns = TE.columns_)
cleaned_df
```

```python
for col in cleaned_df.columns:
    print(col)
```

```python
#dropping any empty columns
df_cleaned = cleaned_df.drop(['nan'], axis =1)
```

```python
#columns after dropping the empty ones
for col in df_cleaned.columns:
    print(col)
```

```python
#Apriori
ap = apriori(df_cleaned, min_support = 0.02, use_colnames = True)
ap.head()
```

```python
ap_results = list(ap)
ap_results
```

```python
#Association Rules
assoc = association_rules(ap, metric = 'lift', min_threshold = 1)
assoc.head(10)
```

```python
#Pruning rules
pru = assoc[assoc['confidence']> 0.2]
print('Only {} pru remain'.format(len(pru)))
```

```python
#Sorting rules
sort_rules = assoc.sort_values('lift', ascending=False).head(3)
sort_rules
```
