```python
df = pd.read_csv('D212Task1.csv')
```

```python
#Compare plots
sns.pairplot(df[['Bandwidth_GB_Year', 'Age', 'Income']])
```

```python
import sklearn.cluster as cluster
```

```python
#Perform kmeans with the chosen variables
kmeans = cluster.KMeans(n_clusters=3, init="k-means++")
kmeans = kmeans.fit(df[['Bandwidth_GB_Year', 'Income']])
```

```python
#K Means fitting
kmeans.cluster_centers_
```

```python
#Add labels to original db
df['Clusters'] = kmeans.labels_
```

```python
#Clusters appear at end of db
df.head()
```

```python
#Customer count per cluster

df['Clusters'].value_counts()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = cluster.KMeans(n_clusters=k)
    km.fit(df[['Bandwidth_GB_Year', 'Income']])
    sse.append(km.inertia_)
```

```python
#Elbow plot to check for accuracy
plt.title('Elbow Plot Optimal K Selection')
plt.xlabel('Cluster(K)')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
#According to elbow plot, we can choose 2 or 3 clusters
```

```python
#Perform Silhouette Score to improve accuracy
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4]
silhouette_avg = []
for num_clusters in range_n_clusters:
    
#Initiate kmeans
 kmeans = cluster.KMeans(n_clusters=num_clusters)
 kmeans.fit(df[['Bandwidth_GB_Year', 'Income']])
 cluster_labels = kmeans.labels_
 #silhouette score
 silhouette_avg.append(silhouette_score(df[['Bandwidth_GB_Year', 'Income']], cluster_labels))
plt.plot(range_n_clusters,silhouette_avg,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

```python
#According to the Silhouette Analysis, the optimal number of k is 3 and the score is 0.56
#A score closer to +1 is more favored over a score closer to -1
```
